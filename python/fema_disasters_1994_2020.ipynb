```python
import pandas as pd

df = pd.read_csv('fema_disasters_1994_2020.csv')

df.head(10)
```

```python
incident_ids = df['femaDeclarationString']
incident_types = df['incidentType']
declaration_types = df['declarationType']
dates = df['incidentBeginDate'] = pd.to_datetime(df['incidentBeginDate']).dt.date

frame = {
    'incident_id': incident_ids,
    'date': dates,
    'declaration_type': declaration_types,
    'incident_type': incident_types
}

df_incidents_by_type = pd.DataFrame(frame)

df_incidents_by_type.to_csv('cleaned_fema_incidents.csv')
```
